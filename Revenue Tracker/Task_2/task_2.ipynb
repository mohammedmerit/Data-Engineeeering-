```python
apr_df = pd.read_excel('Task2_April.xlsx', sheet_name=None, header=None)
may_df = pd.read_excel('Task2_May.xlsx',   sheet_name=None, header=None)
jun_df = pd.read_excel('Task2_June.xlsx',  sheet_name=None, header=None)
out_df = pd.read_excel('Task2_Output.xlsx', sheet_name=None, header=None)

print('April sheets:', list(apr_df.keys()))
print('May sheets:',   list(may_df.keys()))
print('June sheets:',  list(jun_df.keys()))
print('Output sheets:',list(out_df.keys()))
```

```python
apr_df = apr_df['MI - Apr @ 108 ']
may_df = may_df['MI - Apr @ 108 ']
jun_df = jun_df['MI - Apr @ 112.86 ']
out_df = out_df['Revenue ']
```

```python
apr_projects = apr_df.iloc[1, 2:].tolist()  # project names from row 1
apr_values   = apr_df.iloc[3, 2:].tolist()  # revenue values from row 3

may_projects = may_df.iloc[1, 2:].tolist()  # project names from row 1
may_values   = may_df.iloc[3, 2:].tolist()  # revenue values from row 3

jun_projects = jun_df.iloc[1, 2:].tolist()  # project names from row 1
jun_values   = jun_df.iloc[3, 2:].tolist()  # revenue values from row 3

# replace your apr_map, may_map, jun_map lines with these
apr_map = {str(k).strip(): v for k, v in zip(apr_projects, apr_values)}
may_map = {str(k).strip(): v for k, v in zip(may_projects, may_values)}
jun_map = {str(k).strip(): v for k, v in zip(jun_projects, jun_values)}

print(apr_map)
print(may_map)
print(jun_map)
```

```python
from openpyxl import load_workbook

wb = load_workbook('Task2_Output.xlsx')
ws = wb['Revenue ']

# rows 19-28 have project data (openpyxl is 1-indexed, so row 19 = row 19)
for row in range(20, 30):
    project_name = ws.cell(row=row, column=1).value
    if project_name:
        project_name = project_name.strip()  # strip output names too
    
    ws.cell(row=row, column=4).value  = apr_map.get(project_name)
    ws.cell(row=row, column=7).value  = may_map.get(project_name)
    ws.cell(row=row, column=10).value = jun_map.get(project_name)

wb.save('Task2_Output.xlsx')
print('Done!')
```
